```python

# Filtro stringente: SOLO femmine con età maggiore o uguale a 16 anni
df_donne_adulte = df[(df["Sex"] == "female") & (df["Age"] >= 16)].copy()

totale_adulte = len(df_donne_adulte)
sopravvissute_adulte = len(df_donne_adulte[df_donne_adulte["Survived"] == 1])

# Creazione del DataFrame per il grafico
dati_grafico = pd.DataFrame(
    {
        "Stato": ["Totale Donne Adulte", "Sopravvissute Adulte"],
        "Conteggio": [totale_adulte, sopravvissute_adulte],
    }
)

# Impostazione dello stile grafico
sns.set_theme(style="whitegrid")
plt.figure(figsize=(6, 5))

# Creazione del grafico a barre senza warning
ax = sns.barplot(
    x="Stato",
    y="Conteggio",
    hue="Stato",
    data=dati_grafico,
    palette=["#8e44ad", "#27ae60"],
    legend=False,
)

# Configurazione del titolo e delle etichette
plt.title(
    "Rapporto Donne Adulte (Età >= 16): Totale vs Sopravvissute",
    fontsize=13,
    pad=15,
    weight="bold",
)
plt.xlabel("")
plt.ylabel("Numero di Persone", fontsize=11)

# Inserimento dei numeri esatti sopra le barre
for p in ax.patches:
    height = p.get_height()
    ax.annotate(
        f"{int(height)}",
        (p.get_x() + p.get_width() / 2.0, height),
        ha="center",
        va="bottom",
        fontsize=12,
        weight="bold",
        xytext=(0, 3),
        textcoords="offset points",
    )

plt.tight_layout()
plt.show()
```

```python

# 1. Calcolo della dimensione del nucleo familiare (passeggero incluso)
df["FamilySize"] = df["SibSp"] + df["Parch"] + 1

# 2. Categorizzazione in base alla presenza di parenti a bordo
df["Is_Alone"] = df["FamilySize"].apply(
    lambda x: "Persona Singola" if x == 1 else "In Famiglia"
)

# Impostazione dello stile del grafico
sns.set_theme(style="whitegrid")
plt.figure(figsize=(9, 6))

# 3. Creazione del grafico a conteggio (Countplot)
ax = sns.countplot(
    x="Is_Alone",
    hue="Survived",
    data=df,
    palette={0: "#e74c3c", 1: "#2ecc71"},
    order=["Persona Singola", "In Famiglia"],
)

# Configurazione del titolo e delle etichette assi
plt.title(
    "Confronto della Sopravvivenza: Persone Singole vs In Famiglia",
    fontsize=14,
    pad=15,
)
plt.xlabel("Tipologia di Passeggero", fontsize=12)
plt.ylabel("Numero Assoluto di Passeggeri", fontsize=12)

# Configurazione strutturata della legenda
legend_labels = ["Deceduto (0)", "Sopravvissuto (1)"]
plt.legend(title="Esito", labels=legend_labels)

# Aggiunta dei valori numerici sopra le barre per facilitare il calcolo visivo
for p in ax.patches:
    height = p.get_height()
    if height > 0:  # Evita di stampare etichette su barre vuote o nulle
        ax.annotate(
            f"{int(height)}",
            (p.get_x() + p.get_width() / 2.0, height),
            ha="center",
            va="bottom",
            fontsize=10,
            weight="bold",
            xytext=(0, 3),
            textcoords="offset points",
        )

# Mostra il grafico su Anaconda
plt.show()
```

```python
# Creazione del Boxplot corretto senza avvisi
# -> Assegniamo x="Pclass" anche a hue e impostiamo legend=False
ax = sns.boxplot(
    x="Pclass",
    y="Fare",
    hue="Pclass",
    data=df,
    palette="Set2",
    legend=False,
    flierprops={
        "marker": "o",
        "markerfacecolor": "#e74c3c",
        "markersize": 6,
        "markeredgecolor": "none",
    },
)

# Configurazione del titolo e delle etichette degli assi
plt.title(
    "Distribuzione delle Tariffe e Identificazione Outliers per Classe di Viaggio",
    fontsize=14,
    pad=15,
)
plt.xlabel("Classe di Imbarco (Pclass)", fontsize=12)
plt.ylabel("Tariffa del Biglietto (Fare)", fontsize=12)

# Configurazione fissa dei punti di ancoraggio dell'asse X per evitare UserWarning
ax.set_xticks([0, 1, 2])
ax.set_xticklabels(["1° Classe", "2° Classe", "3° Classe"])

# Mostra il grafico
plt.show()
```

```python

# 1. Calcolo e categorizzazione del nucleo familiare
df["FamilySize"] = df["SibSp"] + df["Parch"] + 1
df["Is_Alone"] = df["FamilySize"].apply(
    lambda x: "Persona Singola" if x == 1 else "In Famiglia"
)

# Impostazione dello stile grafico
sns.set_theme(style="whitegrid")
plt.figure(figsize=(11, 7))

# 2. Creazione dello Scatter Plot con Stripplot per evitare la sovrapposizione totale dei punti
# Usiamo 'jitter=True' per distribuire leggermente i punti in larghezza e renderli leggibili
ax = sns.stripplot(
    x="Pclass",
    y="Fare",
    hue="Is_Alone",
    data=df,
    palette={"Persona Singola": "#3498db", "In Famiglia": "#e74c3c"},
    jitter=0.25,
    size=6,
    alpha=0.7,
    order=[1, 2, 3],
)

# Configurazione del titolo e delle etichette degli assi
plt.title(
    "Distribuzione dei Passeggeri: Tariffa per Classe e Stato Familiare",
    fontsize=14,
    pad=15,
    weight="bold",
)
plt.xlabel("Classe di Viaggio (Pclass)", fontsize=12)
plt.ylabel("Tariffa Esatta del Biglietto (Fare)", fontsize=12)

# Configurazione rigorosa dell'asse X e della legenda
ax.set_xticks([0, 1, 2])
ax.set_xticklabels(["1° Classe", "2° Classe", "3° Classe"])
plt.legend(title="Tipologia Passeggero", loc="upper right")

# Mostra il grafico
plt.show()

# 3. Controllo numerico empirico stampato in console per validazione matematica
print("--- VERIFICA NUMERICA DEI PASSEGGERI SOPRA TARIFFA 100 ---")
high_fare_df = df[df["Fare"] > 100]
print(high_fare_df.groupby(["Pclass", "Is_Alone"]).size().to_string())
```

```python
# Rimuoviamo eventuali valori nulli nella colonna Embarked per pulizia grafica
df_clean = df.dropna(subset=["Embarked"]).copy()

# Impostazione della griglia con due grafici affiancati
sns.set_theme(style="whitegrid")
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Palette personalizzata per le classi
colori_classi = {1: "#f1c40f", 2: "#3498db", 3: "#e67e22"}

# --- GRAFICO 1: Distribuzione delle Classi ---
sns.countplot(
    x="Embarked",
    hue="Pclass",
    data=df_clean,
    palette=colori_classi,
    order=["S", "C", "Q"],
    ax=axes[0],
)
axes[0].set_title(
    "Distribuzione delle Classi per Porto di Imbarco",
    fontsize=13,
    weight="bold",
)
axes[0].set_xlabel("Porto di Imbarco", fontsize=11)
axes[0].set_ylabel("Numero di Passeggeri", fontsize=11)
axes[0].set_xticks([0, 1, 2])
axes[0].set_xticklabels(
    ["Southampton (S)", "Cherbourg (C)", "Queenstown (Q)"]
)
axes[0].legend(title="Classe di Viaggio")

# --- GRAFICO 2: Tasso di Sopravvivenza Corretto ---
# Rimuoviamo hue="Embarked" che causava il disallineamento dei dati rispetto alle etichette
sns.barplot(
    x="Embarked",
    y="Survived",
    data=df_clean,
    order=["S", "C", "Q"],
    color="#2ecc71",  # Un unico colore per evitare bug di mappatura sulle etichette
    errorbar=None,
    ax=axes[1],
)
axes[1].set_title(
    "Percentuale di Sopravvivenza per Porto di Imbarco ",
    fontsize=13,
    weight="bold",
)
axes[1].set_xlabel("Porto di Imbarco", fontsize=11)
axes[1].set_ylabel("Tasso di Sopravvivenza (0.0 = 0%, 1.0 = 100%)", fontsize=11)
axes[1].set_xticks([0, 1, 2])
axes[1].set_xticklabels(
    ["Southampton (S)", "Cherbourg (C)", "Queenstown (Q)"]
)

plt.tight_layout()
plt.show()
```
