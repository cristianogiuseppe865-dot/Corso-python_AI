```python
import pandas as pd
from sklearn.datasets import load_wine
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier

# ==========================================
# 1. CARICAMENTO DEL DATASET WINE (PANDAS)
# ==========================================
wine_data = load_wine(as_frame=True)
df = wine_data.frame

# Separiamo le 13 caratteristiche chimiche (X) dalla colonna target (y)
X = df.drop(columns=["target"])
y = df["target"]

# ==========================================
# 2. DIVISIONE DEI DATI IN TRAINING E TEST SET
# ==========================================
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=41
)

# ==========================================
# 3. ESECUZIONE, ADDESTRAMENTO E LOGICA DEI MODELLI
# ==========================================

# --- 1. K-NEAREST NEIGHBORS (KNN) ---
# COME RAGIONA: Ragiona per "vicinato" e analogia geometrica. 
# Quando deve classificare un vino del test set, calcola la distanza matematica 
# (es. Euclidea) tra quel vino e tutti quelli memorizzati nel training set.
# Individua i 5 vini più vicini (n_neighbors=5) e assegna la classe più frequente tra essi.
# NON crea formule: si limita a mappare i punti nello spazio.
model_knn = KNeighborsClassifier(n_neighbors=5)
model_knn.fit(X_train, y_train)
acc_knn = accuracy_score(y_test, model_knn.predict(X_test))


# --- 2. DECISION TREE (Albero di Decisione Singolo) ---
# COME RAGIONA: Ragiona per "scomposizione gerarchica" (domande a cascata).
# Crea una struttura a diagramma di flusso dove ogni nodo pone una domanda netta su una 
# variabile chimica (es. "L'alcol è > 13.5?"). Se SI prende il ramo destro, se NO il sinistro.
# Continua a scendere e a dividere i dati in gruppi sempre più puri finché non giunge 
# a una foglia finale che indica la classe del vino.

model_dt = DecisionTreeClassifier(random_state=5)
model_dt.fit(X_train, y_train)
acc_dt = accuracy_score(y_test, model_dt.predict(X_test))


# --- 3. RANDOM FOREST (Foresta Casuale) ---
# COME RAGIONA: Ragiona per "democrazia e voto di maggioranza".
# Invece di usare un solo albero, ne crea un'intera flotta (qui 100). Ogni albero viene 
# addestrato su una porzione casuale di dati e di caratteristiche chimiche. 
# Quando arriva un nuovo vino, tutti i 100 alberi fanno la propria previsione indipendente. 
# Il modello raccoglie i 100 voti e assegna la classe che ha ottenuto la maggioranza assoluta.

model_rf = RandomForestClassifier(n_estimators=100, random_state=5)
model_rf.fit(X_train, y_train)
acc_rf = accuracy_score(y_test, model_rf.predict(X_test))


# --- 4. REGRESSIONE LOGISTICA ---
# COME RAGIONA: Ragiona per "probabilità statistica".
# Cerca di tracciare una linea di confine (o un iperpiano) tra i dati applicando una formula 
# matematica chiamata "funzione sigmoide". Invece di dire subito la classe, calcola la 
# probabilità percentuale (da 0% a 100%) che un vino appartenga alla cultivar 0, 1 o 2.
# Assegna infine il vino alla classe che ha ottenuto il punteggio di probabilità più alto.

model_lr = LogisticRegression(max_iter=10000, random_state=5)
model_lr.fit(X_train, y_train)
acc_lr = accuracy_score(y_test, model_lr.predict(X_test))


# --- 5. SUPPORT VECTOR MACHINE (SVM) ---
# COME RAGIONA: Ragiona per "massimizzazione dello spazio di sicurezza".
# Il suo obiettivo è tracciare il confine perfetto tra le classi di vino. Non si accontenta 
# di una linea qualunque, ma cerca quella che si trova alla massima distanza possibile 
# dai punti più vicini di ogni classe (chiamati vettori di supporto). Questo corridoio 
# vuoto (il margine) garantisce che il modello rimanga accurato anche con dati futuri.

model_svm = SVC(kernel="linear", random_state=41)
model_svm.fit(X_train, y_train)
acc_svm = accuracy_score(y_test, model_svm.predict(X_test))


# ==========================================
# 4. STAMPA DEL REPORT COMPLETO
# ==========================================
print("-" * 50)
print("REPORT ACCURATEZZA MODELLI (DATASET WINE):")
print("-" * 50)
print(f"1. K-Nearest Neighbors (KNN):        {acc_knn:.2f}")
print(f"2. Decision Tree (Albero Singolo):   {acc_dt:.2f}")
print(f"3. Random Forest:                    {acc_rf:.2f}")
print(f"4. Regressione Logistica:            {acc_lr:.2f}")
print(f"5. Support Vector Machine (SVM):     {acc_svm:.2f}")
print("-" * 50)
```
