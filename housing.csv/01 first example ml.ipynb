```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

# ==========================================
# 1. CARICAMENTO DEL DATASET
# ==========================================
# load_iris() estrae il dataset standard contenente le misurazioni geometriche dei fiori
data = load_iris()

# X memorizza la matrice delle caratteristiche (le 4 feature: lunghezza/larghezza di sepali e petali)
X = data.data

# y memorizza il vettore delle risposte corrette (le etichette target 0, 1, 2 corrispondenti alle specie)
y = data.target

# ==========================================
# 2. DIVISIONE DEI DATI IN SET DI ADDESTRAMENTO E DI TEST
# ==========================================
# Dividiamo i dati per testare l'algoritmo su informazioni che non ha mai visto durante lo studio.
# test_size=0.2 separa il 20% dei dati complessivi da utilizzare come test set per la verifica finale.
# random_state=41 definisce il seme di casualità specifico usato nel tuo screenshot per il campionamento.
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=41
)

# ==========================================
# 3. CREAZIONE DEL MODELLO DI CLASSIFICAZIONE
# ==========================================
# Inizializziamo l'algoritmo Random Forest (Foresta Casuale), un modello basato su un insieme di alberi decisionali.
# n_estimators=100 indica al modello di creare una foresta composta da esattamente 100 alberi indipendenti.
# random_state=41 blocca la casualità interna della generazione degli alberi per garantire risultati costanti.
model = RandomForestClassifier(n_estimators=100, random_state=41)

# ==========================================
# 4. ADDESTRAMENTO DEL MODELLO
# ==========================================
# Il metodo fit avvia il processo di addestramento: i 100 alberi analizzano X_train e imparano a fare previsioni usando y_train
model.fit(X_train, y_train)

# ==========================================
# 5. PREDIZIONE DELLE ETICHETTE PER IL SET DI TEST
# ==========================================
# Il modello addestrato riceve le caratteristiche dei fiori di test (X_test) e ipotizza la specie di ognuno
predictions = model.predict(X_test)

# ==========================================
# 6. CALCOLO E STAMPA DELL'ACCURATEZZA DEL MODELLO
# ==========================================
# accuracy_score calcola la frazione di risposte indovinate confrontando i valori veri (y_test) con quelli previsti (predictions)
accuracy = accuracy_score(y_test, predictions)

# Stampa l'accuratezza formattandola con due cifre decimali (nel tuo caso l'output specifico è 0.90, ovvero il 90%)
print(f"Accuracy: {accuracy:.2f}")
```
